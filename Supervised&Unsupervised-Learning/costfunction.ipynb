```python
import numpy as np
import matplotlib.pyplot as plt

# Training data
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])

# Cost function
def compute_cost(x, y, w, b):
    m = x.shape[0]
    cost_sum = 0

    for i in range(m):
        f_wb = w * x[i] + b
        cost = (f_wb - y[i]) ** 2
        cost_sum += cost

    total_cost = (1 / (2 * m)) * cost_sum
    return total_cost

# Test values
w = 200
b = 100

print("Cost:", compute_cost(x_train, y_train, w, b))
```
